### VADER
* Advantage: Works well for social media text, including emojis and slangs
* Disadvantage: Out of Vocab words are classified as neutral
* https://www.nltk.org/_modules/nltk/sentiment/vader.html
* https://github.com/cjhutto/vaderSentiment#about-the-scoring

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# check for rows where cleaned_text is null
df_isna = df_vader[df_vader.cleaned_text.isna()]
df_isna.head()
```

```python
df_nona = df_vader.dropna(subset=['cleaned_text'], axis=0).copy()
```

```python
def get_vader_sentiment(text):
    """"
    polarity score represents the proportion of texts that falls under this category.
    positive sentiment: compound score >= 0.05
    neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
    negative sentiment: compound score <= -0.05
    """
    return list(analyzer.polarity_scores(text).values())
```

```python
df_nona['neg_sentiment'], df_nona['neu_sentiment'], df_nona['pos_sentiment'], df_nona['compound_sentiment'] = zip(*df_nona['cleaned_text'].map(get_vader_sentiment))
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                     
```
